```
# Importar bibliotecas necesarias
import pandas as pd
import requests
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Configuraci贸n para mejorar la visualizaci贸n de los gr谩ficos
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['figure.dpi'] = 100

# # Extracci贸n
# Definir la URL de la API que contiene los datos
# Basado en la estructura del repositorio de GitHub, asumimos que el archivo JSON est谩 en la rama 'main'
#api_url = "https://raw.githubusercontent.com/alura-cursos/challenge2-data-science-LATAM/main/telecom_churn.json"
api_url = "https://raw.githubusercontent.com/alura-cursos/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json"
#api_url = "https://github.com/alura-cursos/challenge2-data-science-LATAM/blob/main/TelecomX_Data.json"

print(f"Intentando extraer datos de: {api_url}")

try:
    # Realizar la solicitud GET a la API
    response = requests.get(api_url)
    response.raise_for_status()  # Lanza una excepci贸n para errores HTTP (4xx o 5xx)

    # Cargar los datos JSON en un DataFrame de Pandas
    data = response.json()
    df = pd.DataFrame(data)

    print("Datos extra铆dos exitosamente. Primeras 5 filas del DataFrame:")
    print(df.head())
    print(f"\nN煤mero de filas y columnas iniciales: {df.shape}")
    print("\nInformaci贸n del DataFrame:")
    df.info()

except requests.exceptions.RequestException as e:
    print(f"Error al conectar o descargar los datos de la API: {e}")
    df = pd.DataFrame() # Crear un DataFrame vac铆o en caso de error
except ValueError as e:
    print(f"Error al decodificar JSON: {e}")
    df = pd.DataFrame() # Crear un DataFrame vac铆o en caso de error
except Exception as e:
    print(f"Ocurri贸 un error inesperado durante la extracci贸n: {e}")
    df = pd.DataFrame() # Crear un DataFrame vac铆o en caso de error


# # Transformaci贸n
# Realizar limpieza y transformaci贸n de los datos

if not df.empty:
    print("\n# # Transformaci贸n de Datos")

    # Flatten nested JSON columns
    print("\nFlattening nested columns...")
    customer_df = pd.json_normalize(df['customer'])
    phone_df = pd.json_normalize(df['phone'])
    internet_df = pd.json_normalize(df['internet'])
    account_df = pd.json_normalize(df['account'])

    # Concatenate the flattened dataframes with the original dataframe (excluding the original nested columns)
    df = pd.concat([df[['customerID', 'Churn']], customer_df, phone_df, internet_df, account_df], axis=1)

    print("\nDataFrame after flattening nested columns (info):")
    df.info()
    print("\nPrimeras 5 filas del DataFrame despu茅s de aplanar columnas anidadas:")
    print(df.head())

    # 1. Limpiar nombres de columnas: Eliminar espacios y convertir a min煤sculas y renombrar
    print("\nLimpiando y renombrando columnas...")
    # Rename columns with dots before standardizing names
    df.rename(columns={'Charges.Monthly': 'precio_mensual', 'Charges.Total': 'total_gastos',
                       'SeniorCitizen': 'clientes_antiguos', 'tenure': 'antiguedad',
                       'gender': 'genero', 'Partner': 'socios', 'Dependents': 'dependientes',
                       'PhoneService': 'linea_telef贸nica', 'MultipleLines': 'multiples_lineas',
                       'InternetService': 'servicio_internet', 'OnlineSecurity': 'seguridad_online',
                       'OnlineBackup': 'copia_de_seguridad_online', 'DeviceProtection': 'proteccion_dispositivo',
                       'TechSupport': 'soporte_tecnico', 'StreamingTV': 'tv_por_cable',
                       'StreamingMovies': 'streaming_peliculas', 'Contract': 'contrato',
                       'PaperlessBilling': 'pago_factura_electronica', 'PaymentMethod': 'metodo_pago',
                       'customerID': 'customerid', 'Churn': 'cancelacion'}, inplace=True)

    df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
    print("\nNombres de columnas limpiados y renombrados:")
    print(df.columns)

    # 2. Manejo de valores nulos
    print("\nValores nulos antes de la transformaci贸n:")
    print(df.isnull().sum())

    # La columna 'total_gastos' puede contener espacios o valores no num茅ricos
    # Convertir 'total_gastos' a num茅rico. Los errores se convertir谩n a NaN.
    df['total_gastos'] = pd.to_numeric(df['total_gastos'], errors='coerce')

    # Reemplazar valores nulos en 'total_gastos' por la media o mediana
    # Elegimos la mediana para ser menos sensible a valores at铆picos
    median_total_gastos = df['total_gastos'].median()
    df['total_gastos'].fillna(median_total_gastos, inplace=True)
    print(f"\nValores nulos en 'total_gastos' rellenados con la mediana: {median_total_gastos:.2f}")

    # Para 'antiguedad', si hay nulos, podr铆amos rellenarlos con 0 o la media
    # Asumimos que si no tienen antig眉edad, son nuevos o hay un error, as铆 que 0 podr铆a ser un buen indicador
    df['antiguedad'].fillna(0, inplace=True)
    print("\nValores nulos despu茅s de rellenar 'total_gastos' y 'antiguedad':")
    print(df.isnull().sum())

    # 3. Conversi贸n de tipos de datos
    # Convertir 'antiguedad' a entero si no lo es ya
    df['antiguedad'] = df['antiguedad'].astype(int)

    # Convert 'cancelacion' (Churn) column - already renamed to cancelacion
    df['cancelacion'] = df['cancelacion'].map({'Yes': 1, 'No': 0})


    # Las columnas binarias pueden tener 'No internet service' o 'No phone service' que son equivalentes a 'No'
    # 'No' -> 0, 'Yes' -> 1
    # 'No internet service' -> 0 (para 'servicio_internet' y relacionados)
    # 'No phone service' -> 0 (para 'multiples_lineas')
    binary_cols = ['sin_servicio_telefonico', 'multiples_lineas', 'servicio_internet',
                   'seguridad_online', 'copia_de_seguridad_online', 'proteccion_dispositivo',
                   'soporte_tecnico', 'tv_por_cable', 'streaming_peliculas',
                   'pago_factura_electronica'] # Removed 'contrato' and 'cancelacion'

    for col in binary_cols:
        if col in df.columns:
            # Manejar 'No internet service' y 'No phone service' como 'No'
            df[col] = df[col].replace({'No internet service': 'No', 'No phone service': 'No'})
            df[col] = df[col].map({'Yes': 1, 'No': 0})
            # If there are any NaN values after conversion, fill with 0
            df[col].fillna(0, inplace=True)

    # G茅nero: 'Female' -> 0, 'Male' -> 1
    if 'genero' in df.columns and df['genero'].dtype == 'object':
      df['genero'] = df['genero'].map({'Female': 0, 'Male': 1})


    # 'socios' y 'dependientes': 'Yes' -> 1, 'No' -> 0
    if 'socios' in df.columns and df['socios'].dtype == 'object':
      df['socios'] = df['socios'].map({'Yes': 1, 'No': 0})
    if 'dependientes' in df.columns and df['dependientes'].dtype == 'object':
      df['dependientes'] = df['dependientes'].map({'Yes': 1, 'No': 0})

    # 'linea_telef贸nica': 'Yes' -> 1, 'No' -> 0, 'No phone service' -> 0
    if 'linea_telef贸nica' in df.columns and df['linea_telef贸nica'].dtype == 'object':
      df['linea_telef贸nica'] = df['linea_telef贸nica'].replace({'No phone service': 'No'}).map({'Yes': 1, 'No': 0})

    # 'clientes_antiguos' (senior_citizen): Convertir a 0 o 1 si no lo es already numeric
    if 'clientes_antiguos' in df.columns and df['clientes_antiguos'].dtype == 'object':
        df['clientes_antiguos'] = df['clientes_antiguos'].map({'Yes': 1, 'No': 0})
    # It should be numeric after flattening, but keep this check just in case
    elif 'clientes_antiguos' in df.columns:
        df['clientes_antiguos'] = df['clientes_antiguos'].astype(int)


    # 4. Manejo de duplicados
    print(f"\nN煤mero de filas duplicadas antes de la eliminaci贸n: {df.duplicated().sum()}")
    df.drop_duplicates(inplace=True)
    print(f"N煤mero de filas duplicadas despu茅s de la eliminaci贸n: {df.duplicated().sum()}")

    print("\nDataFrame despu茅s de la transformaci贸n (info):")
    df.info()
    print("\nPrimeras 5 filas del DataFrame transformado:")
    print(df.head())
else:
    print("No se pudo realizar la transformaci贸n ya que el DataFrame est谩 vac铆o.")


# # Carga y an谩lisis
# Realizar el an谩lisis exploratorio de datos (EDA)

if not df.empty:
    print("\n# # Carga y An谩lisis")

    # An谩lisis de la variable objetivo 'cancelacion' (churn)
    print("\nDistribuci贸n de clientes que cancelaron (Churn) vs. No Churn:")
    # Ensure 'cancelacion' column exists after flattening and mapping
    if 'cancelacion' in df.columns:
        churn_counts = df['cancelacion'].value_counts()
        churn_percentages = df['cancelacion'].value_counts(normalize=True) * 100
        print(pd.DataFrame({'Conteo': churn_counts, 'Porcentaje': churn_percentages}))

        plt.figure(figsize=(6, 5))
        sns.countplot(x='cancelacion', data=df, palette='pastel')
        plt.title('Distribuci贸n de Cancelaci贸n de Clientes (Churn)')
        plt.xlabel('Cancelaci贸n (0: No, 1: S铆)')
        plt.ylabel('N煤mero de Clientes')
        plt.xticks(ticks=[0, 1], labels=['No Churn', 'Churn'])
        plt.show()
    else:
        print("La columna 'cancelacion' no se encontr贸 en el DataFrame despu茅s de la transformaci贸n.")


    # An谩lisis de variables num茅ricas con respecto a la cancelaci贸n
    numeric_cols = ['antiguedad', 'precio_mensual', 'total_gastos']
    # Filter for numeric columns that actually exist in the DataFrame
    numeric_cols_present = [col for col in numeric_cols if col in df.columns]

    for col in numeric_cols_present:
        plt.figure(figsize=(8, 5))
        sns.histplot(data=df, x=col, hue='cancelacion', kde=True, palette='coolwarm')
        plt.title(f'Distribuci贸n de {col} por Cancelaci贸n')
        plt.xlabel(col)
        plt.ylabel('Frecuencia')
        plt.show()

        plt.figure(figsize=(8, 5))
        sns.boxplot(data=df, x='cancelacion', y=col, palette='coolwarm')
        plt.title(f'{col} por Cancelaci贸n')
        plt.xlabel('Cancelaci贸n (0: No, 1: S铆)')
        plt.ylabel(col)
        plt.xticks(ticks=[0, 1], labels=['No Churn', 'Churn'])
        plt.show()

    # An谩lisis de variables categ贸ricas con respecto a la cancelaci贸n
    categorical_cols = ['genero', 'clientes_antiguos', 'socios', 'dependientes',
                        'linea_telef贸nica', 'multiples_lineas', 'servicio_internet',
                        'seguridad_online', 'copia_de_seguridad_online', 'proteccion_dispositivo',
                        'soporte_tecnico', 'tv_por_cable', 'streaming_peliculas', 'contrato',
                        'pago_factura_electronica', 'metodo_pago']

    for col in categorical_cols:
        if col in df.columns and 'cancelacion' in df.columns:
            plt.figure(figsize=(8, 5))
            sns.countplot(x=col, hue='cancelacion', data=df, palette='viridis')
            plt.title(f'Cancelaci贸n por {col}')
            plt.xlabel(col)
            plt.ylabel('N煤mero de Clientes')
            plt.xticks(rotation=45, ha='right')
            plt.legend(title='Cancelaci贸n', labels=['No Churn', 'Churn'])
            plt.tight_layout()
            plt.show()

            # Calcular el porcentaje de churn para cada categor铆a
            churn_rates = df.groupby(col)['cancelacion'].mean() * 100
            print(f"\nPorcentaje de Cancelaci贸n por {col}:")
            print(churn_rates.sort_values(ascending=False))
        elif col not in df.columns:
            print(f"La columna '{col}' no se encontr贸 en el DataFrame.")


    # Correlaci贸n entre variables num茅ricas
    # Filter for numeric columns that actually exist in the DataFrame
    numeric_cols_present = [col for col in numeric_cols if col in df.columns]
    if len(numeric_cols_present) > 0 and 'cancelacion' in df.columns:
        plt.figure(figsize=(10, 8))
        sns.heatmap(df[numeric_cols_present + ['cancelacion']].corr(), annot=True, cmap='coolwarm', fmt=".2f")
        plt.title('Matriz de Correlaci贸n de Variables Num茅ricas con Cancelaci贸n')
        plt.show()
    elif 'cancelacion' not in df.columns:
        print("La columna 'cancelacion' no se encontr贸 para calcular la correlaci贸n.")
    else:
        print("No hay columnas num茅ricas presentes para calcular la correlaci贸n.")


else:
    print("No se pudo realizar el an谩lisis ya que el DataFrame est谩 vac铆o.")


# # Informe final
# En esta secci贸n, resumiremos las conclusiones clave del an谩lisis exploratorio de datos
# y las posibles razones de la evasi贸n de clientes.

# El objetivo es proporcionar un informe conciso para el equipo de ciencia de datos.
```
